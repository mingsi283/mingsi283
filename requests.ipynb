```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://nba.hupu.com/')
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
items = soup.find_all('div',class_="name")

for item in items:
    print(item.getText())
    print('\n')
```
